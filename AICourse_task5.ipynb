# Prompt 2:
I would like to use the load function, with as_supervised=True and with_info=true

```python
# Import necessary libraries
import tensorflow as tf
import tensorflow_datasets as tfds

# Load the Fashion MNIST dataset with info
dataset, info = tfds.load('fashion_mnist', as_supervised=True, with_info=True)

# Split the dataset into training and testing sets
train_dataset, test_dataset = dataset['train'], dataset['test']

# Display dataset information
print(info)

# Function to normalize images
def normalize_img(image, label):
    return tf.cast(image, tf.float32) / 255.0, label

# Apply normalization to the datasets
train_dataset = train_dataset.map(normalize_img)
test_dataset = test_dataset.map(normalize_img)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to display a grid of images
def display_images(dataset, num_images=25, grid_size=(5, 5)):
    # Take a random sample of images
    sample_images = dataset.shuffle(buffer_size=1000).take(num_images)

    # Create a grid for displaying the images
    plt.figure(figsize=(8, 8))
    for i, (image, label) in enumerate(sample_images):
        plt.subplot(grid_size[0], grid_size[1], i + 1)
        plt.imshow(image.numpy(), cmap='gray')
        plt.title(f"Label: {label.numpy()}")
        plt.axis('off')

    plt.show()

# Display 25 random images in a 5x5 grid
display_images(train_dataset)
```

## Display Distribution as a Pie Chart

```python
def get_counts(dataset):
  # Calculate counts using a loop
    counts = [0] * 10
    for _, label in dataset:
        counts[label.numpy()] += 1
    return counts


labels = [str(i) for i in range(10)]  # 10 classes in Fashion MNIST

train_counts = get_counts(train_dataset)
test_counts = get_counts(test_dataset)
```

```python
# Function to display distribution of clothing items as a pie chart
def display_distribution_pie_chart(counts, labels, title):
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.show()

# Display distribution of clothing items in training set
display_distribution_pie_chart(train_counts, labels, "Distribution of Clothing Items in Training Set")

# Display distribution of clothing items in test set
display_distribution_pie_chart(test_counts, labels, "Distribution of Clothing Items in Test Set")
```

```python
# Function to check if a dataset is balanced
def check_balance(counts, dataset_name):
    min_count = min(counts)
    max_count = max(counts)
    balance_threshold = 0.2  # Adjust this threshold based on your criteria

    if (max_count - min_count) / len(counts) <= balance_threshold:
        print(f"{dataset_name} is balanced.")
    else:
        print(f"{dataset_name} is not balanced.")

# Check balance for training set
check_balance(train_counts, "Training Set")

# Check balance for test set
check_balance(test_counts, "Test Set")
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Assuming train_dataset is a TensorFlow Dataset
# Replace 'features' and 'labels' with the actual names of your features and labels
features = train_dataset.map(lambda x, y: x)  # Assuming your dataset is a tuple (features, labels)
labels = train_dataset.map(lambda x, y: y)

# Convert features and labels to NumPy arrays
features = list(features.as_numpy_iterator())
labels = list(labels.as_numpy_iterator())

# Split the data into training and validation sets
features_train, features_val, labels_train, labels_val = train_test_split(
    features, labels, test_size=0.2, random_state=42
)

# Create TensorFlow Datasets from NumPy arrays
train_dataset = tf.data.Dataset.from_tensor_slices((features_train, labels_train))
val_dataset = tf.data.Dataset.from_tensor_slices((features_val, labels_val))
```

# Building the network
### Prompt:
Today you're my AI coding assistant. We are going to be working to the Fashion-MNIST dataset, and training an AI model to classify from different clothing items.
I have already prepared all the needed code to get the dataset, normalize it, and split into train and test lists.
Now we need to build the model itself. We are working in Google colab and writing our code in Python. We are using TensorFlow with the Keras library.
I need you to build CNN (convolutional neural network). The input layer has a shape of (28,28,1). I need you to build some Conv2D layers, and a MaxPool layer after each Conv2D layer. Afterwards, we need to flatten our data into a vector, and build a couple fully-connected (Dense) layers. The network will have an output layer of 10 neurons (one for each class), with each neuron holding the probability of the corresponding class appearing in the image. We will make our prediction to be the neuron with the highest probability. Our optimizer will be Adam, our loss function will be Cross-Entropy, and our metric is accuracy. On the output layer, our activation function will be SoftMax.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential()

# Convolutional layers with MaxPooling
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Flatten the data
model.add(layers.Flatten())

# Fully connected layers
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))

# Output layer
model.add(layers.Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Display the model summary
model.summary()
```

```python
# Define the number of epochs and batch size
epochs = 10
batch_size = 16

# Train the model on the Fashion MNIST dataset
history = model.fit(train_dataset.batch(batch_size), epochs=epochs, validation_data=val_dataset.batch(batch_size))

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_dataset.batch(batch_size))
print(f"\nTest Accuracy: {test_accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Take a sample of 25 random images from the test dataset
sample_images = test_dataset.shuffle(buffer_size=1000).take(25)

# Function to display images in a grid with correct and predicted classes
def display_predictions(model, dataset, class_names):
    plt.figure(figsize=(15, 15))

    for i, (image, label) in enumerate(dataset):
        # Run the image through the model to get predictions
        predictions = model.predict(np.expand_dims(image, axis=0))
        predicted_class = np.argmax(predictions)

        # Plot the image
        plt.subplot(5, 5, i + 1)
        plt.imshow(image.numpy(), cmap='gray')
        plt.title(f"True: {class_names[label.numpy()]}\nPredicted: {class_names[predicted_class]}")
        plt.axis('off')

        # Display 25 images
        if i == 24:
            break

    plt.show()

# Display the images with correct and predicted classes
display_predictions(model, sample_images, class_names)
```
